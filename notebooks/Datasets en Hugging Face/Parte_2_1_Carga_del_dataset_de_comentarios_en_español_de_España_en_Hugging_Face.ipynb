```python
# Imports
!pip install -q datasets
!pip install huggingface_hub
import pandas as pd
import io
from datasets import load_dataset, load_dataset_builder, get_dataset_split_names, load_dataset, concatenate_datasets, DatasetDict, Dataset
from google.colab import files
```

1.- Cogemos el archivo CSV y creamos una estructura Dataset

```python
# Cargar el fichero
def upload_files (index_fields=None):
  uploaded = files.upload()
  for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(
        name=fn, length=len(uploaded[fn])))
    df = pd.read_csv(io.StringIO(uploaded[fn].decode('utf-8')), index_col = index_fields, delimiter="*")
    df = pd.DataFrame(df)
    return df

dataframe = upload_files()
```

```python
# Cargar el DataFrame de Pandas a un DatasetDict
dataset = Dataset.from_pandas(dataframe)
```

2.- Dividimos el dataset en 3 splits y creamos el objeto DatasetDict

```python
# Dividir el dataset en tres splits: train (70 %), validation (10 %) y test (20 %)

# Priero dividir el dataset en dos partes: train (70 %) y test (dict_aux_2: 30 %)
dict_aux_1 = dataset.train_test_split(test_size=0.3, seed=42)
# Ahora dividir el resto en dos: validation (33 %) y test (66 %)
dict_aux_2 = dict_aux_1['test'].train_test_split(test_size=0.66, seed=42)
```

```python
# Formamos ordenadamente el objeto DatasetDict
dataset_dict = DatasetDict({
    'train': dict_aux_1['train'], # 70 %
    'validation': dict_aux_2['train'], # 10 %
    'test': dict_aux_2['test'] # 20 %
})
```
