```python
!apt-get update
!apt-get install chromium chromium-driver
!pip3 install selenium

import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from google.colab import files
```

```python
service = Service("/usr/bin/chromedriver")
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=service, options=options)
```

```python
# Noticia específica
driver = webdriver.Chrome(service=service, options=options)
info_dict = []
try:
  link="https://theobjective.com/espana/2024-02-18/sanchez-donado-marruecos-vigilar-fronteras/"
  driver.get(link)
  wait = WebDriverWait(driver, 3)
  time.sleep(1)
  title_xpath = "//h1"
  title = wait.until(EC.visibility_of_element_located((By.XPATH, title_xpath))).text
  print(title)
  print(link)

  button_xpath = "//button[@class='tno-comments-box__load-comments']"
  button = wait.until(EC.element_to_be_clickable((By.XPATH, button_xpath)))
  print(button)
  driver.execute_script("arguments[0].click();", button)
  time.sleep(2)

  moreComments = True
  while(moreComments):
    try:
      moreCommentsButton = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'button[class="tno-button load-more-comments visible"]')))
      driver.execute_script("arguments[0].click();", moreCommentsButton)
      time.sleep(2)
    except Exception as e:
      moreComments = False
      continue

  comentarios = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='tno-comment__content']")))
  print("Total comments: ", len(comentarios))
  for comentario in comentarios:
    info_dict.append({
      'link': link,
      'title': title,
      'comment': comentario.text,
      'racism': 0 # Inicializamos a 0 por defecto
    })
except Exception as e:
  print(e)

driver.quit()

import datetime
fileName = 'comentarios_the_objective_' + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") + '.csv'
df = pd.DataFrame(info_dict)
df.to_csv(fileName, encoding='utf-8-sig', sep='*')

files.download(fileName)
```
